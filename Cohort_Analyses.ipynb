```python
oszlopnevek = ['read_date', 'read_time', 'event', 'country', 'user_id', 'source', 'topic']
read = pd.read_csv('/home/kriszta/DILAN/read.csv', delimiter = ';', header = None, names=oszlopnevek)
read = read[['user_id', 'read_date']]
read
```

```python
read['read_date'] = pd.to_datetime(read['read_date'])
read = read[(read['read_date'] >= '2018-03-01') & (read['read_date'] < '2018-03-31')]
read
```

```python
first_read = pd.read_csv('/home/kriszta/DILAN/read.csv', delimiter = ';', header = None, names=oszlopnevek)
first_read = first_read[first_read['source'].notna()]
first_read = first_read[['user_id', 'read_date']]
first_read = first_read.rename(columns={'user_id': '1st_read_user_id', 'read_date': '1st_read_date'})
first_read
```

```python
first_read['1st_read_date'] = pd.to_datetime(first_read['1st_read_date'])
first_read = first_read[(first_read['1st_read_date'] >= '2018-03-01') & (first_read['1st_read_date'] < '2018-03-31')]
first_read
```

```python
oszlopnevek2 = ['subs_date', 'subs_time', 'event', 'user_id']
subscribe = pd.read_csv('/home/kriszta/DILAN/subscribe.csv', delimiter = ';', header = None, names=oszlopnevek2)
subscribe = subscribe[['user_id','subs_date']]
subscribe = subscribe.rename(columns={'user_id': 'subs_user_id', 'subs_date': 'subs_date'})
subscribe
```

```python
subscribe['subs_date'] = pd.to_datetime(subscribe['subs_date'])
subscribe = subscribe[(subscribe['subs_date'] >= '2018-03-01') & (subscribe['subs_date'] < '2018-03-31')]
subscribe
```

```python
oszlopnevek3 = ['buy_date', 'buy_time', 'event', 'user_id', 'price']
buy = pd.read_csv('/home/kriszta/DILAN/buy.csv', delimiter = ';', header = None, names=oszlopnevek3)
buy = buy[['user_id','buy_date']]
buy = buy.rename(columns={'user_id': 'buy_user_id', 'buy_date': 'buy_date'})
buy
```

```python
buy['buy_date'] = pd.to_datetime(buy['buy_date'])
buy = buy[(buy['buy_date'] >= '2018-03-01') & (buy['buy_date'] < '2018-03-31')]
buy
```

### FIRST_READ & READ COHORT

```python
cohort_first_read_read = first_read.merge(read, how = 'inner', left_on = '1st_read_user_id', right_on = 'user_id')
cohort_first_read_read
```

```python
user_read_dates = cohort_first_read_read.groupby('user_id')['read_date'].apply(lambda x: pd.unique(x).tolist())
user_read_dates = user_read_dates.tolist()
user_read_dates = [[pd.Timestamp(x).strftime('%Y-%m-%d') for x in user_dates] for user_dates in user_read_dates]
user_read_dates
```

```python
date_list = ['2018-03-01', '2018-03-02', '2018-03-03', '2018-03-04', '2018-03-05', '2018-03-06', '2018-03-07', 
             '2018-03-08', '2018-03-09', '2018-03-10', '2018-03-11', '2018-03-12', '2018-03-13', '2018-03-14', 
             '2018-03-15', '2018-03-16', '2018-03-17', '2018-03-18', '2018-03-19', '2018-03-20', '2018-03-21', 
             '2018-03-22', '2018-03-23', '2018-03-24', '2018-03-25', '2018-03-26', '2018-03-27', '2018-03-28', 
             '2018-03-29', '2018-03-30', '2018-03-31']

print("cohort;days_passed;number_of_users")
for j in range(15):
    cohort = date_list[j]
    day_passed = 0
    for i in range(j,j+11):
        given_date = date_list[i]
        num_of_users = 0
        for user in user_read_dates:
            if given_date in user and user[0] == cohort:
                num_of_users += 1
        print(cohort, day_passed, num_of_users)
        day_passed += 1
```
